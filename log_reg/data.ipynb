```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([0, 1, 2, 4, 5, 8, 10, 11, 13, 14])
y = np.array([0, 0, 1, 0, 0, 1, 1, 1, 0, 1])
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X, y, s=100)
plt.yticks([0, 0.5, 1])
plt.ylabel("Ders geçme", fontsize=18)
plt.xlabel("Çalışılan saat", fontsize=18)
plt.xlim(-2, 20)
plt.tight_layout()
# plt.savefig("sinav.pdf")
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
z = np.linspace(-10, 10, 100)
y_z = sigmoid(z)

plt.figure(figsize=(10, 5))
plt.plot(z, y_z)
plt.yticks([0, 0.5, 1])
plt.ylabel(r"$\sigma(z)$", fontsize=18)
plt.xlabel(r"$z$", fontsize=18)
plt.tight_layout()
# plt.savefig("sigmoid.pdf")
plt.show()
```

```python
X_log = np.linspace(0.001, 1, 100)
y_log = -np.log(X_log)

plt.figure(figsize=(6, 6))
plt.plot(X_log, y_log)
plt.ylabel(r"$-\log(h_{\theta}(x))$", fontsize=18)
plt.xlabel(r"$h_{\theta}(x)$", fontsize=18)
plt.title(r"$y = 1$", fontsize=18)
plt.tight_layout()
# plt.savefig("log_cost_1.pdf")
plt.show()
```

```python
X_log = np.linspace(0, 0.99, 100)
y_log = -np.log(1 - X_log)

plt.figure(figsize=(6, 6))
plt.plot(X_log, y_log)
plt.ylabel(r"$-\log(1-h_{\theta}(x))$", fontsize=18)
plt.xlabel(r"$h_{\theta}(x)$", fontsize=18)
plt.title(r"$y = 0$", fontsize=18)
plt.tight_layout()
# plt.savefig("log_cost_0.pdf")
plt.show()
```

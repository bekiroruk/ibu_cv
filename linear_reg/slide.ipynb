```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-4, 5)
```

```python
plt.figure(figsize=(12, 4))
fd = {"size": 18}

fig_idx = 1

theta_0_values = [0, 2, -2]
theta_1_values = [1, 2, 0.5]

for theta_0, theta_1 in zip(theta_0_values, theta_1_values):
    y = theta_0 + theta_1 * x

    plt.subplot(1, 3, fig_idx)
    plt.title(rf"$\theta_0: {{{theta_0}}}, \theta_1: {{{theta_1}}}$", fontdict=fd)
    plt.xlim(-6, 6)
    plt.ylim(-6, 6)
    plt.grid()
    plt.plot(x, y)

    fig_idx += 1

plt.tight_layout()
plt.savefig("t0_t1.pdf")
plt.show()
```
